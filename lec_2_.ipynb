### Exercise
Consider the list x = [1.0, 2.0, 3.0]. Write the code to compute the twice of each value and store it in y

```python
x = [1.0, 2.0, 3.0]

y = []
```

```python
for var in x:
  y.append(var*2)
print(y)
```

```python
plt.plot(x, y)
plt.scatter(x, y)
plt.xlabel('number')
plt.ylabel('double')
plt.title('Double Plot')
plt.show();
```
