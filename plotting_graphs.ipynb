```python
# import the matplot library
import matplotlib.pyplot as plt
import numpy as np
```

### Example - 1: first lets learn to plot a simple line plot

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x, y)
plt.show();
```

### Example 2 : Now lets learn how to add the labels
#### for this purpose we will use the xlabel() and ylabel()

```python
plt.plot(x, y)
plt.xlabel('numbers')
plt.ylabel('squares of the numbers')
plt.show();
```

### Example - 3: Now lets try to add a title to our graph
#### We can do that using a title()

```python
plt.plot(x, y)
plt.xlabel('numbers')
plt.ylabel('squares of the numbers')
plt.title('Squares')
plt.show();
```

```python
### Example 4 : Lets see how we can change the line style and color
plt.plot(x, y, color = 'red', linestyle = '--', marker = 'o')
plt.xlabel('numbers')
plt.ylabel('squares')
plt.title('Squares of the numbers')
plt.show()
```

```python
plt.plot(x, y, color = 'blue', linestyle = '-.', linewidth = 2, marker = 'o')
plt.show()
```

### Example - 5 : Now lets see how we can add multiple lines

```python
plt.plot(x, [z**2 for z in x], label='squares')
plt.plot(x, [z**3 for z in x], label='cubes')
plt.legend()
plt.show();
```

### Example - 6 : Gridlines

```python
plt.plot(x, y)
plt.grid(True)
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Squares of the numbers')
plt.show();
```

### Example - 7 : Lets learn how to plot a scatter plot

```python
plt.plot([2,5,45,56], [23,43,54,67])
plt.scatter([2,5,45,56], [23,43,54,67])
plt.show();
```

### Example - 8 : Lets learn how to create a bar chart

```python
plt.bar(['Mumbai','Banglore','Chennai','Delhi'], [250, 450, 57, 87])
plt.xlabel('Cities')
plt.ylabel('Average Salaries')
plt.title('Average Salaries of the cities')
plt.show();
```

### Example 9 : Histogram

```python
data = np.random.randn(1000)

# np.random.randn generates samples from the standard normal distribution:
#	Mean (μ) = 0
#	Standard deviation (σ) = 1
#	1000 means you’re generating 1000 data points.

plt.hist(data, bins = 30)
# plt.hist() creates a histogram — a way of showing how data is distributed.
# How it works internally:
# Bins divide the range of data into intervals.
# Here, bins=30 means the range from min(data) to max(data) is divided into 30 equal-width segments.

plt.show();
```

```python
data = np.random.randn(3000)
plt.hist(data, bins = 23)
plt.show();
```

### Example 5 : We add annotations and legends along with setting limits to the axes

```python
x = np.linspace(-5*np.pi, 5*np.pi, 500)


plt.plot(x, x, color = 'blue')
plt.plot(x, -x, color = 'blue')
plt.plot(x, np.sin(x), color = 'green', linewidth = 2)
plt.plot(x, x*np.sin(x), color = 'red',linewidth = 3)
plt.legend(['x','-x','sin(x)','xsin(x)'], loc = 'upper right')
plt.annotate('origin', xy = (0,0))


plt.xlim(-5*np.pi,5*np.pi)
plt.ylim(-5*np.pi,5*np.pi)

plt.show()
plt.savefig('homework_plot.png');
```

### Example 7 : Now lets learn Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
y_true = [0,1,2,3,4]
y_pred = [1,3,4,5,6]

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt = 'd', cmap = 'Reds')
plt.show();
```

### Example 8 : Loss Curve during training

```python
epochs = range(1, 11)
loss = [0.9, 0.8, 0.7, 0.55, 0.45, 0.32, 0.43, 0.22, 0.99, 0.85]
plt.plot(epochs, loss, marker = 'o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Curve')
plt.show();
```
